```python
# .info() prints information including the index dtype, column dtypes, non-null values, and memory usage

data.info()
```

```python
# .describe() provides basic description and statistics for each column, such as count, mean, standard deviation for
# all numerical values


data.describe()
```

```python
# Returns the data types for each 

data.dtypes
```

```python
# Missing data percentage list
for col in data.columns:
    pct_missing = np.mean(data[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
# According to this there is no missing data, which matches the comments on Kaggle
```

```python
# .nunique() counts the number of unique elements in an object
# I am surprised by how many times the hashtags column has repeated information, given that instagram posts have up to 10 unique hashtags each. 

data.nunique()
```

```python
# I used a box plot to see if there are any potential outliers there may be. (possibly in impressions and from explore)
# This also shows the distribution of numbers in each column

data.plot.box()
```

```python
# This graph is VERY hard to read, but it does show one important thing that I would like to explore further:there is one set of hashtags that is used very frequently


sns.histplot(data['Hashtags'])
plt.show()
```

```python
# Based on the above graph, I wanted to look into how frequently certain groups of hashtags occured, and identify the one that occurs more frequently than the rest
# getting the count of how many times a hashtag group occurred 
data['Hashtags'].value_counts()
```

```python
# I wanted to see if the captions column would have similar results to the hashtags column
# Once again, very hard to read, but does show that there are multiple repeats

sns.histplot(data['Caption'])
plt.show()
```
